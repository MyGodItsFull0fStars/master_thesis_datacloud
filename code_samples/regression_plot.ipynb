# How to Create a Scatterplot with a Regression Line in Python
source: https://www.statology.org/scatterplot-with-regression-line-python/

```python
import numpy as np 

#create data
x = np.array([x for x in range(0, 100, 10)])
y = np.array([1, 1, 1, 2, 3, 4, 4, 5, 6, 7])
len(x), len(y)
```

```python
import matplotlib.pyplot as plt
import random

# create basic scatterplot
plt.plot(x, y, 'o')
plt.ylabel('Number of Cats')
plt.xlabel('Age (yrs)')

# obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x, y, 1)
# y = np.concatenate((y, np.array([x for x in range(0, 7)])))

# add linear regression line to scatterplot 
x_scatter = [x for x in range(0, 100, 10)]
y_scatter = np.random.randint(2, 8, 10)
plt.scatter(x_scatter, y_scatter)
plt.plot(x, m*x+b)
```
